## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline     
#display all figures and return to the ipython prompt.
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

### Functional Method

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
#  1st arg for rows, 2nd arg for columns, 3rd arg for plot no. 1st
plt.subplot(1,2,1)   
plt.plot(x,y,'r')

#  1st arg for rows, 2nd arg for columns, 3rd arg for plot no. 2nd
plt.subplot(1,2,2)  
plt.plot(y,x,'b')
```

### Object-Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])   
# 20% from left, 50% frokm the bottom, 40% to the canvas size and height
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])    

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

'''for current_axes in axes:
        current_axes.plot(x,y) ''' 

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()   # checks for overlapping
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))  
# dpi is Dot's per Inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

#### Setting Labels and Titles

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')
```

#### Adding Legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')
#legend is used to mark the difference in the lines in rhe same graph
ax.legend(loc=(0.1,0.1))  #loc 10 for centre , loc 0 for best fit
```

# Plot Apperance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=2,alpha=0.5) # RGB Hex Code #FF8C00(orange)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=1,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor="red")   # (--,-.,:,steps) linestyle(ls)  marker(+,o,*,1)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])

ax.set_ylim([0,2])
```
